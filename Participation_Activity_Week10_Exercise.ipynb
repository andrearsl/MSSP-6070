```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
 #Andrea has 2 siblings.
G.add_node('Andrea')
G.add_nodes_from(['Nia', 'Noel'])

G.add_edge('Andrea', 'Nia')
G.add_edges_from([('Nia', 'Noel'), ('Andrea', 'Noel')])

plt.figure(figsize=(8, 7.5))
nx.draw_networkx(G)
```

```python
#Andrea's siblings, Nia and Noel has friends; Brianna and Dion, and Andrea's friend, Tony.
G.add_edges_from([('Nia', 'Brianna'), ('Noel', 'Dion'), ('Andrea','Tony')])
nx.draw_networkx(G)
```
